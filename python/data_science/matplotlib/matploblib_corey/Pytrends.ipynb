```python
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
pytrends = TrendReq(hl = 'en-US')
keywords = ['Python', 'JavaScript', 'Java', 'HTML', 'C++']
pytrends.build_payload(keywords, timeframe = 'today 5-y')
```

```python
data = pytrends.interest_over_time()
data
```

```python
plt.plot(data)
plt.suptitle('Interest in Programming Languages(2018-2022)')
plt.xlabel('Years')
plt.ylabel('Level of Interest')
plt.legend(keywords, loc = 'best')
plt.tight_layout()
plt.show()
```

```python
focus = ['JavaScript', 'Python', 'C++']
plt.plot(data[focus])
plt.legend(focus)
```
