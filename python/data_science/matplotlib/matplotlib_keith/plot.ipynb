```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.xkcd
```

```python
gas = pd.read_csv('gas_prices.csv')
#plt.figure(figsize = (8,5))

plt.plot(gas['Year'], gas['USA'], 'b.-', label = 'USA')
plt.plot(gas['Year'], gas['Canada'], 'g.-', label = 'Canada')

plt.legend(bbox_to_anchor = (1.2, 1.2))

plt.title('Gas Prices (USD) over Time')

plt.xticks(gas['Year'][::3]) #list of comprehension
plt.xlabel('Year')
plt.ylabel('US Dollar')
plt.show()
```

```python
for country in gas:
    if country != 'Year':
        plt.plot(gas['Year'], gas[country], marker = '.')
plt.title('Gas Prices (USD) over Time')
plt.xticks(gas['Year'][::3]) #list of comprehension
plt.xlabel('Year')
plt.ylabel('US Dollar')
plt.show()
```

```python
ctrlist = ['USA', 'Canada']
for country in gas:
    if country in ctrlist:
        plt.plot(gas['Year'], gas[country], marker = '.')
plt.title('Gas Prices (USD) over Time', fontdict = {'fontname':'Brush Script MT','fontweight': 'bold', 'fontsize' : '32'})
plt.xticks(gas['Year'][::3].to_list() + [2011]) #list of comprehension
plt.xlabel('Year')
plt.ylabel('US Dollar')
#plt.savefig('Gas_prices.png', dpi = 300)
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
pd.set_option('display.max_columns', 89)
fifa.head()
```

```python
bins = [30, 40, 50, 60, 70, 80, 90, 100]
plt.hist(fifa['Overall'], bins = bins, edgecolor = 'black', color = '#abcdef')
plt.xticks(bins)
plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player skills in FIFA 2018')

plt.show()
```

```python
labels = ['Right', 'Left']
explode = [0, 0.1]
plt.pie(fifa['Preferred Foot'].value_counts(), explode = explode, labels = labels, autopct = '%.1f%%',
        shadow = True, counterclock = False, wedgeprops = {'ec': 'black'})
plt.show()
```

```python
fifa['Weight'] = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight
```

```python
xhv_filt = fifa['Weight'] >= 200
hv_filt = (fifa['Weight'] > 170) & (fifa['Weight'] < 200)
md_filt = (fifa['Weight'] <= 170) & (fifa['Weight'] >= 160)
l_filt = (fifa['Weight'] < 160) & (fifa['Weight'] >= 125)
xl_filt = fifa['Weight'] < 125
xlight = fifa.loc[xl_filt].count()[0]
light = fifa.loc[l_filt].count()[0]
medium = fifa.loc[md_filt].count()[0]
heavy = fifa.loc[hv_filt].count()[0]
xheavy = fifa.loc[xhv_filt].count()[0]
```

```python
plt.style.use('bmh')
labels = ['Under 125lbs','125-160lbs', 'Medium Weight(160-170lbs)', '170-200lbs', 'Over 200lbs']
slices = [xlight, light, medium, heavy, xheavy]
explode = [0.5, 0.1, 0, 0 , 0.4]
plt.pie(slices, explode = explode, labels = labels, startangle = 0, autopct = '%.1f%%')
plt.title('Weight Distribution of FIFA players (in lbs)')
plt.show()
```

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

filt = fifa['Club'] == 'FC Barcelona'
filt2 = fifa['Club'] == 'Real Madrid'
filt3 = fifa['Club'] == 'Manchester United'

barca = fifa.loc[filt, 'Overall']
madrid = fifa.loc[filt2, 'Overall']
mu = fifa.loc[filt3, 'Overall']
labels = ['FC Barcelona', 'Real Madrid', 'Manchester United']
plt.boxplot([barca, madrid, mu], labels = labels)
plt.title('Professional Football Team Comparison')
plt.ylabel('FIFA Overall Rating')
plt.show()
```

```python
boxes = plt.boxplot([barca, madrid, mu], labels = labels, patch_artist = True,
                    medianprops = {'linewidth': 2}) #Patch_artist when wanna change facecolor
color1 = ['#42b3f5', '#a7f542', '#42f59e']
color2 = ['#748279','#78abaa','#8c78ab']
for box in boxes['boxes']:
    #Edge color
    box.set(color = '#42b3f5', linewidth = 2)
    #Fill color
    box.set(facecolor = '#e0e0e0')
    
plt.show()
    
```
